```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000
y = np.random.uniform(0, 1, N) #gerar N amostras uniformes

x = y**(2/9) #Gerar as amostras de x

fx = (1 - x**5)**(7/2) # Avaliação da função original f(x)
gx = (9/2) * x**(7/2)   # Avaliação da função de importância g(x) nos pontos x

integral = np.sum(fx / gx) / N # Calcular a integral usando o estimador correto
print(f'Integral = {integral:.4f}')

#plotar o grafico
x_plot = np.linspace(0, 1, N)
curve_y = (1 - x_plot**5)**(7/2)
plt.plot(x_plot, curve_y, color='black')
plt.fill_between(x_plot, curve_y, color='blue')
plt.text(0.5, 0.8, f'Area = {integral:.4f}', fontsize=14, color='red')
plt.xlim(-0.01, 1.1)
plt.ylim(-0.01, 1.1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Integrando f(x)')
plt.show()
```
