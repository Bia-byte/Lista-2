```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000  # número de amostras

# Gerar amostras x diretamente da distribuição de importância g(x)
u = np.random.uniform(0, 1, N)
x = u / np.sqrt(1 - u**2)

# f(x) = x^2 / (1 + x^2)^3
fx = (x**2) / (1 + x**2)**3

# g(x) = (1 + x^2)**(-3/2)
gx = (1 + x**2)**(-3/2)

# Estimativa da integral usando importância: f(x) / g(x)
integral = np.mean(fx / gx)
print(f'Integral = {integral:.4f}')

# Plotar o gráfico (adaptado para a variável x)
x_plot = np.linspace(0, 10, 500)
curve_y = x_plot**2 / (1 + x_plot**2)**3
plt.plot(x_plot, curve_y, color='black')
plt.fill_between(x_plot, curve_y, color='blue')
plt.text(2, 0.04, f'Area = {integral:.4f}', fontsize=14)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integrando f(x)')
plt.show()
```
