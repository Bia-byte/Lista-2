```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000  # número de amostras

# Gerar amostras y a partir da distribuição g(y) = 3y^2 com y ∈ [0, 1]
# inversa da CDF: x = (3u/A)^(1/3) com A = 3
u = np.random.uniform(0, 1, N)
A = 3
y = ((3 * u) / A) ** (1/3)

# Calcular x a partir de y: x = (1 - y)/y
x = (1 - y) / y

# f(x) = x^2 / (1 + x^2)^3
fx = (x**2) / (1 + x**2)**3

# g(y) = 3 * y^2
gy = A * y**2

# Estimativa da integral usando importância: f(x)/g(y)
integral = np.sum(fx / gy) / N
print(f'Integral = {integral:.4f}')

#plotar o grafico
x = np.linspace(0, 30, 500)
curve_y = x*((1 + x**2)**(-2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(5, 0.3,f'Area = {integral:.4f}', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
